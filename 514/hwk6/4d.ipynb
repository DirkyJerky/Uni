```python
import numpy as np
from numpy.polynomial.chebyshev import Chebyshev
from matplotlib import pyplot as plt
```

```python
xs = np.linspace(-2,3,10001)
```

```python
f = lambda x: np.sin(x)
p1 = lambda x: 0.256*x
rho1 = lambda x: 0.438*x - 0.104
```

```python
plt.plot(xs, f(xs), label='$f(x)$')
plt.plot(xs, p1(xs), label='$p_1(x)$')
plt.plot(xs, rho1(xs), label='$\\rho_1(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$f(x), p_1(x), \\rho_1(x)$')
plt.legend()
plt.show()
```
