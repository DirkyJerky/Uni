```python
# First we import some modules:
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
# And define a function
def f(x):
    f = np.cos(x)**2 - x*np.sin(x)/(1.0+x**2)
    return f
```

```python
# To plot, we must have imported matplotlib (see the first input cell)
# Then...

x = np.linspace(-20,20,1000) # A uniform discretization of [-20,20] using 1000 points

plt.figure()
plt.plot(x,f(x))
plt.plot(x,np.zeros_like(x),'k--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('A function!')
plt.show()
```

```python
# Let's see if we can find a root of f(x) using the simplest iterative method.
# First we define:

def g(x):
    return x-f(x)
```

```python
# Now to seed the iteration with an initial guess and set the number of iterations:

x0 = -10.0 # Initial guess
iterations = 20    
```

```python
# Iteration:

x = x0
plt.figure()

for k in np.arange(iterations):
    x = g(x)
    print('x = ', x)
    print('|x-g(x)| = ', np.abs(x-g(x)))
    plt.plot(k,x,'r.')
plt.xlabel('$k$')
plt.ylabel('$x_k$')
plt.title('Simple iteration')
plt.show()
```
