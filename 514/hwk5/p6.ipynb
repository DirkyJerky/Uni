```python
import numpy as np
from numpy.polynomial.chebyshev import Chebyshev
from matplotlib import pyplot as plt
```

```python
xs = np.linspace(-1,1,10001)
```

```python
for i in range(15):
    plt.plot(xs, Chebyshev.basis(i)(xs), label='$T_{' + str(i) + '}$')

plt.xlabel('$x$')
plt.ylabel('$T_i (x)$')
plt.title('$T_0, T_1, ..., T_{14}$')
plt.show()
```

```python
fx7 = lambda x: x**7
T7 = Chebyshev.basis(7)
minimax = lambda x: fx7(x) - (2**(-6))*T7(x)
```

```python
plt.plot(xs, fx7(xs), label='$x^7$')
plt.plot(xs, minimax(xs), label = '$p_6(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```

```python
plt.plot(xs, fx7(xs) - minimax(xs))
plt.show()
```
